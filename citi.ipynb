```python
def _encode_for_display(text):
    """
    Encodes strings so they can display as ASCII in a Windows terminal window.
    This function also encodes strings for processing by xml.etree.ElementTree functions.

    Returns an ASCII-encoded version of the text.
    Unicode characters are converted to ASCII placeholders (for example, "?").
    """
    return text.encode('ascii', errors="backslashreplace").decode('utf-8')

def login_viz(server_cred):
    '''
    this function takes input as a dictionary of the credentials of the server which you want to login and returns 
    the XSRF-token and response to use the cookies
    '''
    
    REQ_SOURCE_SERVER_URL = server_cred['url']
    REQ_SOURCE_SERVER_USERNAME = server_cred['username']
    REQ_SOURCE_SERVER_PASSWORD = bytes(server_cred['password'],'UTF-8')



    r = requests.post('{}/vizportal/api/web/v1/generatePublicKey'.format(REQ_SOURCE_SERVER_URL), json={"method": "generatePublicKey","params": {}})
    resp = r.json()
    keyPair = resp['result']['key']
    n = int(keyPair['n'],16)
    e = int(keyPair['e'],16)
    pubkey = RSA.construct((n,e))
    msg = REQ_SOURCE_SERVER_PASSWORD
    cipher = PKCS1_v1_5.new(pubkey)
    encrypted_pass = cipher.encrypt(msg)
    x =binascii.b2a_hex(encrypted_pass).decode()
    tableau_username = REQ_SOURCE_SERVER_USERNAME
    tab_server_url = REQ_SOURCE_SERVER_URL
    session = requests.Session()
    def vizportalLogin(encryptedPassword, keyId):
        #  encodedPassword = binascii.b2a_hex(encryptedPassword)
         encodedPassword = encryptedPassword
         payload = "{\"method\":\"login\",\"params\":{\"username\":\"%s\", \"encryptedPassword\":\"%s\", \"keyId\":\"%s\"}}" % (tableau_username, encodedPassword,keyId)
         endpoint = "login"
         url = tab_server_url + "/vizportal/api/web/v1/"+endpoint
         headers = {
         'content-type': "application/json;charset=UTF-8",
         'accept': "application/json, text/plain, */*",
         'cache-control': "no-cache"
         }
         response = session.post(url, data=payload, headers=headers)
         return response
    login_response = vizportalLogin(x, resp['result']['keyId'])
    log_res = login_response.cookies
    a = log_res.get_dict()
    TOk_XSRF = a['XSRF-TOKEN']
    TOk_XSRF

    tab_server_url = REQ_SOURCE_SERVER_URL

    Ex_Json= {
        "method":"getSites",
        "params":
        {
            "order":
            [
                {
                    "field":"name",
                    "ascending":'true'
                }
            ],
            "page":
            {
                "startIndex":'0',
                "maxItems":'100'
            }
        }
    }

    payload = json.dumps(Ex_Json)

    endpoint = Ex_Json['method']
    url = tab_server_url + "/vizportal/api/web/v1/"+endpoint
    headers = {
    'content-type': "application/json;charset=UTF-8",
    'accept': "application/json, text/plain, */*",
    'cache-control': "no-cache",
    'X-XSRF-TOKEN': TOk_XSRF
    }
    cookies = login_response.cookies
    site_response = requests.post(url, data=payload, headers=headers,cookies=cookies)
    site_response 

    sites_url = dict()
    sites_id = dict()

    for i in site_response.json()['result']['sites']:
        sites_url[i['settings']['name']] = i['settings']['urlName']

    for i in site_response.json()['result']['sites']:
        sites_id[i['settings']['name']] = i['id']

    if str(server_cred['site']).lower() != 'default':
        tab_server_url = REQ_SOURCE_SERVER_URL

        Ex_Json= {"method":"switchSite","params":{"urlName":sites_url[server_cred['site']]}}

        payload = json.dumps(Ex_Json)

        endpoint = Ex_Json['method']
        url = tab_server_url + "/vizportal/api/web/v1/"+endpoint
        headers = {
        'content-type': "application/json;charset=UTF-8",
        'accept': "application/json, text/plain, */*",
        'cache-control': "no-cache",
        'X-XSRF-TOKEN': TOk_XSRF
        }
        cookies = login_response.cookies
        ser_response = requests.post(url, data=payload, headers=headers,cookies=cookies)
        TOk_XSRF = ser_response.cookies.get_dict()
        TOk_XSRF = TOk_XSRF['XSRF-TOKEN']
        print(TOk_XSRF)
    else:
        ser_response = login_response

    return TOk_XSRF,ser_response

def get_sheet_type(REQ_SOURCE_SERVER_URL,TOk_XSRF,ser_response):
    tab_server_url = REQ_SOURCE_SERVER_URL
    viz_views_lst = []
    count = 1000
    index = 0 
    while index <= count:
        Ex_Json= {
            "method":"getViews",
            "params":
            {
                "order":
                [
                    {
                        "field":"name",
                        "ascending":'true'
                    }
                ],
                "page":
                {
                    "startIndex":str(index),
                    "maxItems":'1000'
                },
                "feilds":
                [
                    'id',
                    'luid',
                    'sheetType'
                ]
            }
        }

        payload = json.dumps(Ex_Json)

        endpoint = Ex_Json['method']
        url = tab_server_url + "/vizportal/api/web/v1/"+endpoint
        headers = {
        'content-type': "application/json;charset=UTF-8",
        'accept': "application/json, text/plain, */*",
        'cache-control': "no-cache",
        'X-XSRF-TOKEN': TOk_XSRF
        }
        cookies = ser_response.cookies
        response = requests.post(url, data=payload, headers=headers,cookies=cookies)
        count = int(response.json()['result']['totalCount'])
        index += 1000
        res_list = response.json()['result']['views']
        viz_views_lst += res_list
    response

    sheet_type = dict()
    for i in viz_views_lst:
        Ex_Json= {"method":"getView","params":{"id":i['id']}}
        payload = json.dumps(Ex_Json)

        endpoint = Ex_Json['method']
        url = tab_server_url + "/vizportal/api/web/v1/"+endpoint
        headers = {
        'content-type': "application/json;charset=UTF-8",
        'accept': "application/json, text/plain, */*",
        'cache-control': "no-cache",
        'X-XSRF-TOKEN': TOk_XSRF
        }
        cookies = ser_response.cookies
        response = requests.post(url, data=payload, headers=headers,cookies=cookies)
        if 'sheetType' in response.json()['result'].keys():
            sheet_type[response.json()['result']['luid']] = response.json()['result']['sheetType']
    #         print(response.json())
    return sheet_type

def get_total_hit(server_cred,VERSION,site_id,auth_token):
    page_num, page_size = 1, 1000
    views_list = []
    while True:
        url = server_cred['url'] + "/api/{0}/sites/{1}/views?includeUsageStatistics=true".format(VERSION, site_id)
        paged_url = url + "&pageSize={0}&pageNumber={1}".format(page_size, page_num)
        server_response = requests.get(paged_url, headers={'x-tableau-auth': auth_token})
        page_num += 1
        if 'error' not in xmltodict.parse(server_response.text)['tsResponse'].keys():
            views_list.extend(xmltodict.parse(server_response.text)['tsResponse']['views']['view'])
        else:
            break

    total_hit_dict = dict()
    for view in views_list:
        total_hit_dict[view['@id']] = view['usage']['@totalViewCount']
    return total_hit_dict


def login_tableau_tsc(server_cred):
    """
    Checks if their is any data available on the server
    """
    server_url=  server_cred['url']
    username = server_cred['username']            
    password = server_cred['password']                               
    site =  server_cred['site']
    tableau_auth = TSC.TableauAuth(username,password,site)
    server = TSC.Server(server_url,use_server_version=True)
    server.auth.sign_in(tableau_auth)
    return tableau_auth,server

def check_data(server_response):
    """
    Checks if their is any data available on the server
    """
    server_response_dict = xmltodict.parse(server_response.text)
    if int(server_response_dict['tsResponse']['pagination']['@totalAvailable'])==0:
        raise NodataError('No data available on the server')
    return

def _check_status(server_response, success_code):
    """
    Checks the server response for possible errors.

    'server_response'       the response received from the server
    'success_code'          the expected success code for the response
    Throws an ApiCallError exception if the API call fails.
    """
    if server_response.status_code != success_code:
        parsed_response = ET.fromstring(server_response.text)
        xmlns = {'t': 'http://tableau.com/api'}

        # Obtain the 3 xml tags from the response: error, summary, and detail tags
        error_element = parsed_response.find('t:error', namespaces=xmlns)
        summary_element = parsed_response.find('.//t:summary', namespaces=xmlns)
        detail_element = parsed_response.find('.//t:detail', namespaces=xmlns)

        # Retrieve the error code, summary, and detail if the response contains them
        code = error_element.get('code', 'unknown') if error_element is not None else 'unknown code'
        summary = summary_element.text if summary_element is not None else 'unknown summary'
        detail = detail_element.text if detail_element is not None else 'unknown detail'
        error_message = '{0}: {1} - {2}'.format(code, summary, detail)
        raise ApiCallError(error_message)
    return


def login_rest(server_cred):
    url = server_cred['url'] + "/api/{0}/auth/signin".format(VERSION)
    
    # Builds the request
    xml_request = ET.Element('tsRequest')
    credentials_element = ET.SubElement(xml_request, 'credentials', name=server_cred['username'], password=server_cred['password'])
    ET.SubElement(credentials_element, 'site', contentUrl=server_cred['site'])
    xml_request = ET.tostring(xml_request)
    
    # Make the request to server
    xmlns = {'t': 'http://tableau.com/api'}
    server_response = requests.post(url, data=xml_request)
    _check_status(server_response,200)

    # ASCII encode server response to enable displaying to console
    server_response = _encode_for_display(server_response.text)

    # Reads and parses the response
    parsed_response = ET.fromstring(server_response)
    
    xmlns = {'t': 'http://tableau.com/api'}
    
    token = parsed_response.find('t:credentials', namespaces=xmlns).get('token')
    site_id = parsed_response.find('.//t:site', namespaces=xmlns).get('id')
    user_id = parsed_response.find('.//t:user', namespaces=xmlns).get('id')
    
    return token,site_id,user_id

def get_extract_jobs(server_cred,auth_token,site_id,VERSION):
    page_num, page_size = 1, 1000
    
    url = server_cred['url'] + "/api/{0}/sites/{1}/jobs".format(VERSION, site_id)
    paged_url = url + "?pageSize={0}&pageNumber={1}".format(page_size, page_num)
    server_response = requests.get(paged_url, headers={'x-tableau-auth': auth_token})
    
    _check_status(server_response,200)
    check_data(server_response)
    
    server_response_dict = xmltodict.parse(server_response.text)
    
    extract_jobs = []
    for job in server_response_dict['tsResponse']['backgroundJobs']['backgroundJob']:
        if 'refresh_extracts' in job['@jobType'].lower():
            extract_jobs.append(job)
            
    if len(extract_jobs) == 0:
        raise NodataError
        return
    else:
        for job in extract_jobs:
            try:
                job_id = job['@id']
                page_num, page_size = 1, 1000
                url = server_cred['url'] + "/api/{0}/sites/{1}/jobs/{2}".format(VERSION, site_id,job_id)
                paged_url = url + "?pageSize={0}&pageNumber={1}".format(page_size, page_num)
                server_response = requests.get(paged_url, headers={'x-tableau-auth': auth_token})

                job_info = xmltodict.parse(server_response.text)
                notes = job_info['tsResponse']['job']['extractRefreshJob']['notes']
                datasource = job_info['tsResponse']['job']['extractRefreshJob']['datasource']
                job['@notes'] = notes
                job['@datasource'] = datasource

            except:
                job['@notes'] = 'no info for '+job['@id']
        return extract_jobs

def get_project_details(tableau_auth):
    with server.auth.sign_in(tableau_auth):
        all_projects = []
        cnt = 1000
        page = 1
        while cnt == 1000:
            a_w,p_i = server.projects.get(req_options=TSC.RequestOptions(pagenumber=page,pagesize=1000))
            cnt = len(a_w)
            page += 1
            all_projects += a_w
            
        project_df = pd.DataFrame()
        
        for prj in all_projects:
            project_df = pd.concat([project_df,pd.DataFrame({'project_id':[prj.id],'project_name':[prj.name]})],ignore_index=True)
            
    return project_df

def get_datasource_details(tableau_auth):
    with server.auth.sign_in(tableau_auth):
        all_datasources = []
        cnt = 1000
        page = 1
        while cnt == 1000:
            a_w,p_i = server.datasources.get(req_options=TSC.RequestOptions(pagenumber=page,pagesize=1000))
            cnt = len(a_w)
            page += 1
            all_datasources += a_w

        datasource_df = pd.DataFrame()

        for ds in all_datasources:
            datasource_df = pd.concat([datasource_df,pd.DataFrame({'datasource_id':[ds.id],
                                                                   'datasource_name':[ds.name],
                                                                   'owner_id':[ds.owner_id],
                                                                   'project_id':[ds.project_id]})],ignore_index=True)
    return datasource_df

def get_workbook_details(tableau_auth):
    with server.auth.sign_in(tableau_auth):
        all_workbooks = []
        cnt = 1000
        page = 1
        while cnt == 1000:
            a_w,p_i = server.workbooks.get(req_options=TSC.RequestOptions(pagenumber=page,pagesize=1000))
            cnt = len(a_w)
            page += 1
            all_workbooks += a_w

        workbook_df = pd.DataFrame()

        for wb in all_workbooks:
            workbook_df = pd.concat([workbook_df,pd.DataFrame({'workbook_id':[wb.id],
                                                               'workbook_name':[wb.name],
                                                               'owner_id':[wb.owner_id],
                                                               'project_id':[wb.project_id]})],ignore_index=True)
    return workbook_df

def get_group_details(tableau_auth):
    with server.auth.sign_in(tableau_auth):
        all_groups = []
        cnt = 1000
        page = 1
        while cnt == 1000:
            a_w,p_i = server.groups.get(req_options=TSC.RequestOptions(pagenumber=page,pagesize=1000))
            cnt = len(a_w)
            page += 1
            all_groups += a_w

        groups_df = pd.DataFrame()

        for group in all_groups:
            server.groups.populate_users(group)
            for user in group.users:
                groups_df = pd.concat([groups_df,pd.DataFrame({'group_id':[group.id],
                                                               'group_name':[group.name],
                                                               'user_id':[user.id]})],ignore_index=True)
    return groups_df

def get_user_details(tableau_auth):
    with server.auth.sign_in(tableau_auth):
        all_user = []
        cnt = 1000
        page = 1
        while cnt == 1000:
            a_w,p_i = server.users.get(req_options=TSC.RequestOptions(pagenumber=page,pagesize=1000))
            cnt = len(a_w)
            page += 1
            all_user += a_w

        user_df = pd.DataFrame()

        for user in all_user:
            user_df = pd.concat([user_df,pd.DataFrame({'user_id':[user.id],
                                                       'user_name':[user.name],
                                                       'user_role':[user.site_role]})],ignore_index=True)
    return user_df
```
